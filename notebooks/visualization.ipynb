```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# 1. Correct Pathing: Go up one level from /notebooks to reach /data
data_path = '../data/processed/sentiment_results.csv'

if os.path.exists(data_path):
    df = pd.read_csv(data_path)
    print("✅ Data loaded successfully!")
    
    # Set the visual style
    sns.set_theme(style="whitegrid")
    plt.rcParams['figure.figsize'] = [12, 6]

    # --- CHART 1: Sentiment Distribution per Bank ---
    plt.figure()
    sns.countplot(data=df, x='bank', hue='sentiment_label', palette='viridis')
    plt.title('Sentiment Distribution by Bank (CBE vs BOA vs Dashen)', fontsize=15)
    plt.ylabel('Number of Reviews')
    plt.show()

    # --- CHART 2: Confidence vs. Star Rating (Aggregation) ---
    plt.figure()
    # Using a boxplot to show the spread of AI confidence for each star
    sns.boxplot(data=df, x='rating', y='sentiment_score', palette='Spectral')
    plt.title('AI Sentiment Confidence Spread per Star Rating', fontsize=15)
    plt.xlabel('User Star Rating')
    plt.ylabel('DistilBERT Confidence Score')
    plt.show()

    # --- CHART 3: Top Pain Points (Themes) ---
    plt.figure()
    # Filter for only negative reviews to see what's wrong
    neg_reviews = df[df['sentiment_label'] == 'NEGATIVE']
    sns.countplot(data=neg_reviews, y='identified_theme', hue='bank', palette='magma')
    plt.title('Identify Pain Points: Negative Themes per Bank', fontsize=15)
    plt.xlabel('Number of Complaints')
    plt.ylabel('Thematic Category')
    plt.show()

else:
    print(f"❌ Error: Could not find file at {data_path}")
    print("Ensure you have run 'python scripts/analyze_reviews.py' in your terminal first.")
```
